```python
import math

print("Using math module to perform mathematical operations")
print("The square root of 16 is:", math.sqrt(16)) #prints 4.0
print(math.floor(3.7)) #prints 3
print(math.ceil(3.3))  #prints 4
```

```python
a = 5
b = 2
c = a / b
print("The division of a and b is:", c) #float division
print("The data type of c is:", type(c)) #float

print(5.0/2) #float
print(5/2.0) #float
print(5.0/2.0) #float   


c = a%b
print("The modulus of a and b is:", c) #modulus
print("The data type of c is:", type(c)) #integer
c = a//b
print("The floor division of a and b is:", c) #floor division
print("The data type of c is:", type(c)) #integer

c = a**b
print("The exponentiation of a and b is:", c) #exponentiation   
print("The data type of c is:", type(c)) #integer

c = 7
print("Is c between 0 and 10?:", 0 < c < 10) #True
```

```python
#strings
s = "Hello, World!"
print("The string is:", s)
print("the length of the string is:", len(s)) #length of string
print("the first character of the string is:", s[0]) #first character
print("the last character of the string is:", s[-1]) #last character
print("the substring from index 0 to 4 is:", s[0:5]) #substring
print("the string in uppercase is:", s.upper()) #uppercase
print("the string in lowercase is:", s.lower()) #lowercase
print("the string after replacing 'world' with 'python' is:", s.replace("world", "python")) #replace
print("the string after splitting by comma is:", s.split(",")) #split by comma result is a  list of hello and world
print("is 'hello' in the string?:", "hello" in s) #true
print("is 'python' not in the string?:", "python" not in s) #true   
print("the string after stripping whitespace is:", "   hello   ".strip()) #strip removes leading and trailing whitespace
print("the string after formatting is:", "hello, {}!".format("alice")) #formatting will replace {} with alice
print("the string after f-string formatting is:", f"hello, {'bob'}!") #f-string formatting will replace {'bob'} with bob
#print("the index of 'w' in the string is:", s.index("w")) #index of character
print("the count of 'l' in the string is:", s.count("l")) #count of character
print("the string after centering is:", s.center(30, "-")) #centering the string with '-' padding   
print("the string after justifying left is:", s.ljust(30, "-")) #left justify with '-' padding
print("the string after justifying right is:", s.rjust(30, "-")) #right justify with '-' padding
print("the string after encoding is:", s.encode("utf-8")) #encoding the string
print("the string after decoding is:", s.encode("utf-8").decode("utf-8")) #decoding the string
print("the string after checking is alphanumeric?:", s.isalnum()) #false because of space and punctuation
print("the string after checking is alphabetic?:", s.isalpha()) #false because of space and punctuation
print("the string after checking is digit?:", s.isdigit()) #false because of letters
print("the string after checking is lowercase?:", s.islower()) #false because of uppercase letters
print("the string after checking is uppercase?:", s.isupper()) #false because of lowercase letters
print("the string after checking is title case?:", s.istitle()) #true because first letter of each word is uppercase
print("the string after checking is space?:", s.isspace()) #false because of letters
print("the string after checking starts with 'hello'?:", s.startswith("hello")) #true
print("the string after checking ends with 'world!'?:", s.endswith("world!")) #true
print("the string after finding 'o' is at index:", s.find("o")) #index of first occurrence of character
print("s[0:16:2]", s[0:16:2]) #slicing with step of 2
print("s[0::-1]", s[0::-1]) #slicing with negative step gives first character
print("s[:16:-1])", s[:16:-1]) #empty string because step is negative and no start
print("s[0:16:-1]", s[0:16:-1]) #empty string because step is negative and start is less than end
print("s[16:0:-1]", s[16:0:-1]) #reverse string without first character...interesting but even if i use -1 we don't get full reversal
print("s[16:-1:-1]", s[16:-1:-1]) #empty string because step is negative and end is negative
print("s[16::-1]", s[16::-1]) #reverse string .. better to give no 0 to get full reversal
print("s[16:0:1]", s[16:0:1]) #got empty string because step is positive and start is greater than end
print("s[::-1]", s[::-1]) #reversing the string

print("String concatenation:", s + " Welcome to Python programming.")
print("String repetition:", s * 3)
print("String with escaped characters: Hello,\nWorld!\tWelcome to Python.")
print("Raw string: Hello,\\nWorld!\\tWelcome to Python.")
print("Multiline string:\n'''Hello,\nWorld!\nWelcome to Python.'''")
print("String with Unicode characters: Hello, \u2602 \u2603")
print("String with byte literals:", b"Hello, World!")
s1 = "??"
print(s,s1)
```

```python
#lists
a = [1, 2, 3, 4, 5] #list
print("The list is:", a)
print("The first element of the list is:", a[0]) #first element
print("The last element of the list is:", a[-1]) #last element
print("The length of the list is:", len(a)) #length of list
a.append(6) #append element
print("The list after appending 6 is:", a)
print(','.join(str(x) for x in a)) #print the list as a string
a.insert(0, 0) #insert element at index 0
print("The list after inserting 0 at index 0 is:", a)
a.insert(1, 1.5) #insert element at index 1
print("The list after inserting 1.5 at index 1 is:", a)
print(a.remove(3)) #remove element 3
print("The list after removing 3 is:", a)
print(a.pop()) #remove last element
print("The list after popping last element is:", a)
a.sort(reverse=True) #sort the list in descending order
print("The list after sorting in descending order is:", a)
a.sort() #sort the list in ascending order
print("The list after sorting in ascending order is:", a)
print("The index of 4 in the list is:", a.index(4)) #index of element
print("The count of 2 in the list is:", a.count(2)) #count of element
a.reverse() #reverse the list
print("The list after reversing is:", a)
print("The sublist from index 1 to 3 is:", a[1:4]) #sublist
print(len(a)) #length of list
print(a*2) #repeat the list
a.extend([7, 8, 9]) #extend the list
print("The list after extending with [7, 8, 9] is:", a) #modified list
print("The concatenated list by not changing the orignal list:", a+[7,8,9]) #just printing the concatenated list without modifying original list
print("List a",a);

for i in a:
    print(i*2)
    print("inside for loop")
print("outside for loop") #prints after for loop completion, indentation matters in python

```

```python
print(a)
#for loop with range()
for i in range(0,10): #increment by 1 default
    print(i)
    print("inside for loop")
#for loop with range() with step and accessing list elements
for i in range(0,len(a),2): #increment by 2
    print(i)
    print(a[i])
    print("inside for loop")
#for loop with list slicing so increasing step by 2
for i in a[::2]:
    print(i)
    print("inside for loop")
```
